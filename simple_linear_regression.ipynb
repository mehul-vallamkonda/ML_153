```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('housing_area_price.csv')
df
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
new_df = df.drop('price',axis='columns')
new_df
```

```python
price = df.price
price
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df,price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_[0]
```

```python
reg.intercept_
```

```python
fig, ax = plt.subplots(figsize=(8,6))

ax.scatter(df['area'], df['price'])

# Define parameters of the regression line
num = len(df)
start = df['area'].min()
end = df['area'].max()
xseq = np.linspace(start, end, num=num)

# Plot the line
ax.plot(xseq, reg.intercept_+reg.coef_[0]*xseq, color="black", lw=1.5)

# Add a title and axis label
ax.set_title('Linear regression: area and price')
ax.set_xlabel('area')
ax.set_ylabel('price')

# Show the plot
plt.show()
```

### For canada housing price

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head()
```

```python
new_df = df.drop('per capita income (US$)',axis='columns')
new_df.head()
```

```python
income = df['per capita income (US$)']
income.head()
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df,income)
```

```python
fig, ax = plt.subplots(figsize=(8,6))

ax.scatter(df['year'], df['per capita income (US$)'])

# Define parameters of the regression line
num = len(df)
start = df['year'].min()
end = df['year'].max()
xseq = np.linspace(start, end, num=num)

# Plot the line
ax.plot(xseq, reg.intercept_+reg.coef_[0]*xseq, color="black", lw=1.5)

# Add a title and axis label
ax.set_title('Linear regression: area and price')
ax.set_xlabel('year')
ax.set_ylabel('per capita income')

# Show the plot
plt.show()
```
