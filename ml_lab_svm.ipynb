```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()

digits.target

df = pd.DataFrame(digits.data,digits.target)
df.head()

df['target'] = digits.target

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('target',axis='columns'), df.target, test_size=0.3)

"""Using RBF kernel"""

from sklearn.svm import SVC
rbf_model = SVC(kernel='rbf')

len(X_train)

len(X_test)

rbf_model.fit(X_train, y_train)

print(rbf_model.score(X_test,y_test))

"""Using Linear kernel"""

linear_model = SVC(kernel='linear')
linear_model.fit(X_train,y_train)

linear_model.score(X_test,y_test)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Load the dataset
iris_data = pd.read_csv('iris.csv')

# Assuming the last column is the target variable (species)
X = iris_data.drop(columns=['species']) # Features
y = iris_data['species'] # Target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Create and train the linear kernel SVM
linear_svm = SVC(kernel='linear')
linear_svm.fit(X_train, y_train)

# Create and train the RBF kernel SVM
rbf_svm = SVC(kernel='rbf')
rbf_svm.fit(X_train, y_train)
```

```python
# Predict on the test set using both models
linear_predictions = linear_svm.predict(X_test)
rbf_predictions = rbf_svm.predict(X_test)

# Calculate accuracy scores
linear_accuracy = accuracy_score(y_test, linear_predictions)
rbf_accuracy = accuracy_score(y_test, rbf_predictions)

print("Linear Kernel Accuracy:", linear_accuracy)
print("RBF Kernel Accuracy:", rbf_accuracy)
```
